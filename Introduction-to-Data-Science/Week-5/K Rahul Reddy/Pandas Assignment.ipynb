```python
import numpy as np
import pandas as pd
Match_data = pd.read_csv("DIM_PLAYER_MATCH.csv", encoding = "ISO-8859-1", delimiter = ',', skiprows = [1])
Match_data.head()
```

```python
young = Match_data[Match_data.Age_As_on_match < 25][['Player_Id','Player_Name', 'Batting_hand', 'Bowling_skill', 'Season_year', 'Player_team', 'is_manofThematch']]
```

```python
data = Match_data[['Player_Id','Player_Name', 'Batting_hand', 'Bowling_skill', 'Season_year', 'Player_team', 'is_manofThematch']]
```

```python
percent = []
team = []
for year in range(2008,2018):
    team += [young[young.Season_year == year].Player_team.value_counts().index[0]]
    percent += [len(young.Player_Id[young.Season_year == year].unique())/len(data.Player_Id[data.Season_year == year].unique())*100]
```

```python
pd.DataFrame([percent, team],columns = data.Season_year.unique(), index = ['Percentage of young players', 'Team with maximum young players '])
```

```python
percent_MOM = len(young[young.is_manofThematch == 1])/len(data[data.is_manofThematch == 1])*100
print("Percent of young players winning Man Of the Match =", percent_MOM)
```

```python
if len(Match_data.Batting_hand.value_counts()) == 5:
    Match_data.Batting_hand.loc[Match_data.Batting_hand.str.contains('Right')] = 'Right'
    Match_data.Batting_hand.loc[Match_data.Batting_hand.str.contains('Left')] = 'Left'
Match_data.Batting_hand.value_counts().sum()
```

```python
if len(Match_data.Bowling_skill.value_counts()) == 20:
    Match_data.Bowling_skill.loc[Match_data.Bowling_skill.notnull() == False] = 'None'
    Match_data.Bowling_skill.loc[Match_data.Bowling_skill.str.contains('Right')] = 'Right'
    Match_data.Bowling_skill.loc[Match_data.Bowling_skill.str.contains('eft')] = 'Left'
    Match_data.Bowling_skill.loc[Match_data.Bowling_skill.str.contains('break')] = 'None'
Match_data.Bowling_skill.value_counts().sum()
```

```python
if len(Match_data.Bowling_skill.value_counts()) == 3:
    Match_data.Bowling_skill.loc[Match_data.Bowling_skill.str.contains('None')] = Match_data[Match_data.Bowling_skill.str.contains('None')]['Batting_hand'] 
```

```python
Right_Bat = Match_data[Match_data.Batting_hand == 'Right']
Right_Bowl = Match_data[Match_data.Bowling_skill == 'Right']
Left_Bat = Match_data[Match_data.Batting_hand == 'Left']
Left_Bowl = Match_data[Match_data.Bowling_skill == 'Left']
Right = pd.concat([Right_Bat,Right_Bowl])
Left = pd.concat([Left_Bat, Left_Bowl])
print("Percentage of Left Players winning MOM =", len(Left[Left.is_manofThematch ==1])/len(Left)*100)
print("Percentage of Right Players winning MOM =", len(Right[Right.is_manofThematch == 1])/len(Right)*100)
print("Left Players have performed better")
```

```python
teams = []
Win = []
Loss = []
for team in M.Player_team.value_counts().index:
    teams.append(team) 
    Win.append(max(M[M.IsPlayers_Team_won == 1][M.Player_team == team]['Opposit_Team'].value_counts()//11))
    Loss.append(max(M[M.IsPlayers_Team_won == 0][M.Player_team == team]['Opposit_Team'].value_counts()//11))             
result = pd.DataFrame([Win, Loss], columns = teams, index = ['Max wins against one team', 'Max losses against one team'])
result
```
